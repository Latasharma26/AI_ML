```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_excel('area_price.xlsx')
df
```

```python
#plot a scatter plot just to get a distribution of data
# %matplolib inline 
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area,df.Price,color='red',marker='+')
```

```python
reg =linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[3300]])
```

```python
reg.intercept_
```

```python
reg.coef_
```
