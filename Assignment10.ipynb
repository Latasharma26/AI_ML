```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_excel('sleep.xlsx')
df
```

```python
plt.xlabel('Hours of Sleep')
plt.ylabel('Hours of Study')
plt.scatter(df['Hours of Sleep'], df['Hours of Study'], color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Hours of Study']], df['Hours of Sleep'])
```

```python
reg.predict([[33]])
```

```python
c=reg.intercept_
c
```

```python
m=reg.coef_
m
```

```python
# y = mx + c
y = m * df['Hours of Study'] + c
print(y)
```

```python

df['Predicted Sleep'] = y
```

```python
print("Predicted Sleep:")
print(df[['Hours of Study', 'Predicted Sleep']])
```
